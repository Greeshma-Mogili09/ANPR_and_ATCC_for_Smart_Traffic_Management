```python

def load_and_display_images(dataset_dir):
    
    if not os.path.exists(dataset_dir):
        print("Dataset directory not found.")
        return
    
    
    files = [f for f in os.listdir(dataset_dir) if os.path.isfile(os.path.join(dataset_dir, f))]
    
    if not files:
        print("No files found in the dataset directory.")
        return
    
    for file in files:
        
        file_path = os.path.join(dataset_dir, file)
       
        image = cv2.imread(file_path)
        
        if image is None:
            print(f"Could not read image {file}")
            continue
   
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        
        plt.figure()
        plt.imshow(image_rgb)
        plt.title(file)
        plt.axis('off')
        plt.show()

load_and_display_images(dataset_dir)
```

```python
import cv2
import os

dataset_dir = r"C:\Users\mgree\Downloads\Images_Dataset_im - Copy"
output_dir = r"C:\Users\mgree\Downloads\Images"


os.makedirs(output_dir, exist_ok=True)


def process_dataset(dataset_dir, output_dir):
    if not os.path.exists(dataset_dir):
        print("Dataset directory not found.")
        return

 
    files = [f for f in os.listdir(dataset_dir) if os.path.isfile(os.path.join(dataset_dir, f))]
    if not files:
        print("No images found in the dataset.")
        return

    for file in files:
        file_path = os.path.join(dataset_dir, file)
        image = cv2.imread(file_path)
        if image is None:
            print(f"Error: Could not load image {file}")
            continue

        print(f"Processing image: {file}")

  
        font = cv2.FONT_HERSHEY_SIMPLEX  
        font_scale = 1  
        font_color = (255, 255, 255) 
        thickness = 2 
        position = (10, 30)  

        text_size, _ = cv2.getTextSize(file, font, font_scale, thickness)
        text_width, text_height = text_size
        background_start = (position[0] - 5, position[1] - text_height - 5)
        background_end = (position[0] + text_width + 5, position[1] + 5)
        cv2.rectangle(image, background_start, background_end, (0, 0, 0), -1) 
        cv2.putText(image, file, position, font, font_scale, font_color, thickness, cv2.LINE_AA)

      
        cv2.imshow("Image with File Name", image)
        print("Press any key to view the next image...")
        cv2.waitKey(0) 
        cv2.destroyAllWindows()

        output_path = os.path.join(output_dir, file)
        cv2.imwrite(output_path, image)
        print(f"Saved image to: {output_path}")

process_dataset(dataset_dir, output_dir)
```
